### Liczby, tekst (string) i zmienne

```python
# Zapisywanie zmiennych
x = 5           # zmienna typu integer
y = 3.14        # zmienna typu float
z = "hello"     # zmienna typu string (text)
q = True        # zmienna typu boolean (logic)
```

```python
z
```

```python
# Głowne operacje
a = 5 + 2       # dodawanie
b = 5 - 2       # odejmowanie
c = 5 * 2       # multiplication
d = 5 / 2       # division
e = 5 % 2       # modulus
f = 5 ** 2      # exponentiation

print(f'a = {a}')
print(f'b = {b}')
print(f'c = {c}')
print(f'd = {d}')
print(f'e = {e}')
print(f'f = {f}')
```

```python
# ZADANIE
# a równe jest 9
# b równe jest 51
# Zapisz a i b jako zmienne i umieść w następującym równaniu:
# Ile wynosi pierwiastek z a pomnożony przez b, a następnie całość spotęgowana przez 3?
```

```python
# Sprawdzanie zależności (comparison operators)
g = 5 == 2      # równe (equal to)
h = 5 != 2      # nierówne (not equal to)
i = 5 > 2       # większe niż (greater than)
j = 5 < 2       # mniejsze niż (less than)
k = 5 >= 2      # większe lub rónwe (greater than or equal to)
l = 5 <= 2      # mniejsze lub rónwe (less than or equal to)

print(f'g = {g}')
print(f'h = {h}')
print(f'i = {i}')
print(f'j = {j}')
print(f'k = {k}')
print(f'l = {l}')
```

```python
# Logiczne operatory
m = True and False      # logiczne AND
n = True or False       # logiczne OR
o = not True            # logiczne NOT
print(f'm = {m}')
print(f'n = {n}')
print(f'o = {o}')
```

```python
# Conditionale (if, elif, else)
if z == 2.5:
    print('"z" jest równe 2.5')
elif z == 5:
    print('"z" jest równe 5')
else:
    print('"z" ma inną wartość')

# Printujemy pustą linię dla celów wizuanych
print('\n')

# f-stringi pozwalają printować wartość jakiejś zmiennej
if y>=x:
    print(f'y równe {y} jest WIĘKSZE od x lub RÓWNE x, które wynosi {x}')
else:
    print(f'y równe {y} jest MNIEJSZE od x, które wynosi {x}')
```

### Listy, indeksowanie i loop (pętla)

```python
# W pythonie jednym z podstawowych obiektów jest lista - może zawierać różne typy obiektów w środku, np. tekst, liczby i wszystkie obiekty
my_list = ['a', 2, [x,y], 4,5 ]
my_list
```

```python
# Możemy napisać pętlę, która pozwoli nam wyświetlać każdy element listy
# najprostrsze pętle to "for x in y" i "while x<y"

for item in my_list: #zamiast "item" może być cokolwiek, np "list_element", "i"
    print('Kolejny element listy:')
    print(item)
```

```python
# Można też napisać pętle generującą liczby
# Używając while
i = 1
while i<=5:
    print(i)
    i=i+1
```

```python
# Uzywając for
for i in range(5):
    print(i)
```

```python
# Lista (i wiele innych obiektów) posiada indeks dla każdego z elementów.
# Możemy wybrać konkretny element używając []
# UWAGA: W Pythonie pierwszą cyfrą indeksu jest prawie zawsze 0
my_list[0]
my_list[2]
# Wybierając zmienne bez printa, w jupyterze wyświetli się tylko ostatnia niestety, dlatego warto pamiętać o princie
```

```python
print(my_list[0])
print(my_list[2])
print(my_list[-1]) #-1, -2 itd będzie wybierało ostatnie elementy listy
```

```python
# Możemy też wybrać kawałek (slice) z listy używając [od : do(bez tego elementu)]
print(my_list[1:3]) # wszystko od elementu 1 do 2 włacznie (bez 3) -> 1, 2
print(my_list[:3]) # elementy do 2 włącznie (bez 3) -> 0, 1, 2
print(my_list[3:]) # elementy od 3 włącznie do końca -> 4, 5
```
